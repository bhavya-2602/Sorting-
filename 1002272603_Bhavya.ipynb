```python
import time
import random
import matplotlib.pyplot as plt
```

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

def selection_sort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr
```

```python
sizes = [5, 10, 20, 50, 100, 200, 500, 1000, 2000, 5000]
results = {'Insertion Sort': [], 'Selection Sort': [], 'Bubble Sort': []}

for size in sizes:
    arr = [random.randint(0, 1000) for _ in range(size)]
    for name, func in [('Insertion Sort', insertion_sort), ('Selection Sort', selection_sort), ('Bubble Sort', bubble_sort)]:
        arr_copy = arr.copy()
        start_time = time.perf_counter()
        func(arr_copy)
        end_time = time.perf_counter()
        results[name].append(end_time - start_time)

for name, times in results.items():
    plt.plot(sizes, times, label=name)

plt.xlabel('Array Size')
plt.ylabel('Time (seconds)')
plt.title('Sorting Algorithm Performance')
plt.legend()
plt.show()
```

# Performance Analysis

The plotted results clearly show the time complexity differences between the three algorithms. Insertion Sort and Selection Sort both exhibit O(n^2) behavior but differ slightly in practical scenarios due to their inherent operations, with Selection Sort generally performing better on larger datasets due to fewer overall data movements. Bubble Sort, although also O(n^2), tends to perform worse because of the number of swaps required.

This benchmark not only validates the theoretical complexities but also provides practical insights into how the algorithms scale with increasing data sizes.
